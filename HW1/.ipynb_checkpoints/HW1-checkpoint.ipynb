```python
finland = pd.Series(df[years][df.country == 'Finland'].to_numpy()[0])
usa = pd.Series(df[years][df.country=='United States'].to_numpy()[0])
france = pd.Series(df[years][df.country == 'France'].to_numpy()[0])
brazil = pd.Series(df[years][df.country == 'Brazil'].to_numpy()[0])
```

```python
plt.figure(figsize=(16,10))

data_plot = pd.DataFrame({"Finland":finland, "USA":usa,"France":france,"Brazil":brazil},)


lp = sns.lineplot(data=data_plot)
lp.set(xlabel='Years',ylabel='Counts')
lp.set_xticklabels([0,1980,1985,1990,1995,2000,2005,2010,2015,2020])

plt.show()
```

```python
count = df.dtypes.value_counts() 

labels = []

for i in count.index:
    labels.append(str(i))

values = []
for value in count:
    values.append(value)
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(32, 12))
#create pie chart
plt.pie(values, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
